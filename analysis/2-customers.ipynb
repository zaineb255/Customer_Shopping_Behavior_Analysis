```python
#missing values
df.isnull().sum()
#to replace missing values its preferrable to use median over mean because mean gets easily affected by outliers
#now we have to be careful with filling the missing value with median and not fill all the missing ones with median because like that we will be
#forcing par exemple the clothing category to get the same rating as footware category which can introduce bias into our data set therfore we are 
#going to fill those missing value with the median of each category seperetly (if we have a missing Review Rating of a clothing item it will be
#replaced by median of Review Rating of clothing item not any other)
```

```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x : x.fillna(x.median()))
#even the fucking space 'Review Rating ' instead of 'Review Rating' can generate a fucking error
```

```python
df.isnull().sum()
```

```python
#its prefferable to not have spaces between column name : Customer ID  or upper and lower cases , soo we convert
df.columns = df.columns.str.lower()  
df.columns = df.columns.str.replace(' ', '_')
df.columns

```

```python
df= df.rename(columns={"purchase_amount_(usd)":"purchase_amount"})
df.columns
```

```python
#create a column age_group
labels= ['young_adult', 'adult', 'middle_aged', 'senior']
df["age_group"]= pd.qcut(df['age'], q=4, labels=labels)
#qcut : splits the age into 4 equal sized group based on the data distribution and assignes the labels we defined
```

```python
df[['age','age_group']].head(20)
```

```python
#NEW COLUMN PURCHASE_FREQUENCY_DAYS
frequency_mapping = {'Fortnightly' : 14, 
                     'Weekly' : 7, 'Monthly': 30, 'Quarterly': 90 , 'Bi-Weekly' : 14 , 'Every 3 Months': 90 , 'Annually' : 365}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['frequency_of_purchases','purchase_frequency_days']].head(20)
```

```python
df[['discount_applied', 'promo_code_used']].head()
#to check if they have the same values or no in this dataset they are the same even though generally a company could have discount applied without
#a promo_code but not in this dataset and here is the proof : 
(df['discount_applied']==df['promo_code_used']).all()
#both columns carry the same infos
```

```python
df = df.drop('promo_code_used' , axis=1)
```
